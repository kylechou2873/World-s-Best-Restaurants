```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd

```

```python
topRestData = pd.read_csv("./Resource/cleaned_Res.csv")
```

```python
topRestData
```

```python
repeatRank = topRestData['restaurant'].value_counts()
repeatRank['Mugaritz']
```

```python
m = folium.Map(location=(12,-5), zoom_start=2)
marker_cluster = MarkerCluster().add_to(m)
reMark = []
for i,row in topRestData.iterrows():
    if row['restaurant'] not in reMark:
        folium.Marker(
            location = [row['lat'],row['lng']],
            popup = f"<h3>{row['restaurant']}</h3><br><p><strong>Times in top 50:</strong> {repeatRank[row['restaurant']]}<br><strong>Region:</strong> {row['Region']}</p>"
        ).add_to(marker_cluster)
        reMark.append(row['restaurant'])
```

```python
m.save("index.html")
```
